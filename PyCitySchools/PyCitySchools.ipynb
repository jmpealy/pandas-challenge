# PyCity Schools Analysis

- The combined dataset spanned 15 schools with over 39,000 students and a total aggregated budget of almost $25 million.  The below analysis looked at average scores in math and reading as well as average passing rates in those categories (70% or higher).  It looked at these statistics by school, by total budget per student, by school type (district vs charter), by grade (10th grade, 11th grade etc) and finally by school size (measured in total students).  Numerous tables are presented throughout the analysis comparing math and reading performance metrics by these different school criteria.  Across the dataset, overall passing rates were nearly 75% in math, and 85% in reading but just 65% for both categories combined.  Average scores and average passing rates generally did not change much by grade (9th, 10th etc) within agiven school.  Small and medium sized schools (less than 1000 students and 1000-2000 students) fared much better than large schools (>2000 students).  Finally, charter schools fared significantly better than district schools.

- From the analysis, two things really stood out to me.  First, total spending per student (budget per capita) was actually negatively correlated with better student performance, which is opposite to what you might normally expect (i.e. higher per student spending produces better results).  Second, despite much better overall performance - charter schools average budget per student was much smaller than that of most district schools.  It's worth mentioning that all but one of the charter schools fell into either the small or medium-sized categories while every one of the district schools fell into the 'large' category.  

- Based on these observations, it would seem that higher spending per student (budget per capita) is not a good predictor of student performance, and if anything could be an indication of school budgets being misspent.  From this dataset, the clearest indicator of strong student performance seems to be school size - the smaller the school the better the students do.  Presumably, this is because smaller schools have smaller classroom size and thus teachers are able to devote more attention to each individual student.

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("../Resources/schools_complete.csv")
student_data_to_load = Path("../Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_df.head()
```

```python
#Rename some columns
school_data_df = school_data_df.rename(columns={"student_name": "Student Name","gender":"Gender","school_name":"School Name"})
school_data_df.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = school_data_df["School Name"].nunique()
school_count
```

```python
# Calculate the total number of students
student_count = school_data_df["Student Name"].count()
student_count
```

```python
# Calculate the total budget
total_budget = school_data_df.groupby("School Name")["budget"].mean().sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_df["math_score"].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_df["reading_score"].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_df[(school_data_df["math_score"] >= 70)].count()["Student Name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_df[(school_data_df["reading_score"] >= 70)].count()["Student Name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_df[(school_data_df["math_score"] >= 70) \
                                                  & (school_data_df["reading_score"] >= 70)].count()["Student Name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
#Define and create a DF for the district summary
district_summary = {"Total Schools": [school_count],
                                "Total Students": [student_count],
                               "Total Budget": [total_budget],
                               "Average Math Score": [average_math_score],
                               "Average Reading Score": [average_reading_score],
                               "% Passing Math": [passing_math_percentage],
                               "% Passing Reading": [passing_reading_percentage],
                               "% Overall Passing": [overall_passing_rate]}
district_summary_df = pd.DataFrame(district_summary)

#Formatting
district_summary_df["Total Schools"] = district_summary_df ["Total Schools"].map("{:,}".format)
district_summary_df ["Total Students"] = district_summary_df ["Total Students"].map("{:,}".format)
district_summary_df ["Total Budget"] = district_summary_df ["Total Budget"].map("${:,.2f}".format)
district_summary_df ["Average Math Score"] = district_summary_df ["Average Math Score"].map("{:.4f}".format)
district_summary_df ["Average Reading Score"] = district_summary_df ["Average Reading Score"].map("{:.4f}".format)
district_summary_df ["% Passing Math"] = district_summary_df ["% Passing Math"].map("{:.4f}%".format)
district_summary_df ["% Passing Reading"] = district_summary_df ["% Passing Reading"].map("{:.4f}%".format)
district_summary_df ["% Overall Passing"] = district_summary_df ["% Overall Passing"].map("{:.4f}%".format)

# Display the DataFrame
district_summary_df
```

## School Summary

```python
# Use the code provided to select all of the school types
school_types = school_data.set_index(["school_name"])["type"]
school_types
```

```python
# Calculate the total student count per school
per_school_counts = school_data_df.groupby("School Name")["Student Name"].count()
per_school_counts
```

```python
# Calculate the total school budget and per capita spending per school
per_school_budget = school_data_df.groupby("School Name")["budget"].mean()
per_student_capita = per_school_budget/per_school_counts
per_student_capita
```

```python
# Calculate the average test scores per school
per_school_math = school_data_df.groupby(["School Name"])["math_score"].mean()
per_school_reading = school_data_df.groupby(["School Name"])["reading_score"].mean()
```

```python
# Calculate the number of students per school with math scores of 70 or higher
students_passing_math = school_data_df[(school_data_df["math_score"] >= 70)]
school_students_passing_math = students_passing_math.groupby(["School Name"]).size()
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
students_passing_reading = school_data_df[(school_data_df["reading_score"] >= 70)]
school_students_passing_reading = students_passing_reading.groupby(["School Name"]).size()
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = school_data_df[
    (school_data_df["math_score"] >= 70) & (school_data_df["reading_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["School Name"]).size()
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = school_students_passing_math / per_school_counts * 100
per_school_passing_reading = school_students_passing_reading / per_school_counts * 100
overall_passing_rate = school_students_passing_math_and_reading / per_school_counts * 100
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = {"School Type": school_types,
                      "Total Students": per_school_counts,
                      "Total School Budget": per_school_budget,
                      "Per Student Budget": per_student_capita,
                      "Average Math Score": per_school_math,
                      "Average Reading Score": per_school_reading,
                      "% Passing Math": per_school_passing_math,
                      "% Passing Reading": per_school_passing_reading,
                      "Overall Passing %": overall_passing_rate}

#Create a per school summary DF
per_school_summary_df = pd.DataFrame(per_school_summary)

#Formatting
per_school_summary_df["Total School Budget"] = per_school_summary_df["Total School Budget"].map("${:,.1f}".format)
per_school_summary_df["Per Student Budget"] = per_school_summary_df["Per Student Budget"].map("${:,.1f}".format)
per_school_summary_df["% Passing Math"] = per_school_summary_df["% Passing Math"].map("{:,.4f}".format)
per_school_summary_df["% Passing Reading"] = per_school_summary_df["% Passing Reading"].map("{:,.4f}".format)
per_school_summary_df["Overall Passing %"] = per_school_summary_df["Overall Passing %"].map("{:,.4f}".format)

# Display the DataFrame
per_school_summary_df
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
buckets = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary_df.copy()

#Change formatting of the 'per student budget' back to int from string
school_spending_df[school_spending_df.columns[1:]] = school_spending_df[school_spending_df.columns[1:]].replace('[\$,]','',regex=True).astype(float)
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"]
                                                             ,spending_bins,labels=buckets
                                                             ,include_lowest=True)
school_spending_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Overall Passing %"].mean()
```

```python
# Assemble into DataFrame
spending_summary = {"Average Math Score": spending_math_scores,
                    "Average Reading Score": spending_reading_scores,
                    "% Passing Math": spending_passing_math,
                    "% Passing Reading": spending_passing_reading,
                    "Overall Passing %": overall_passing_spending}

spending_summary = pd.DataFrame(spending_summary)

# Display results
spending_summary
```

## Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.
per_school_summary = per_school_summary_df.copy()
#Change formatting of the 'per student budget' back to int from string
per_school_summary[per_school_summary.columns[1:]] = per_school_summary[per_school_summary.columns[1:]].replace('[\$,]','',regex=True).astype(float)

per_school_summary["School Size"] = pd.cut(per_school_summary["Total Students"], 
                                           size_bins,labels=labels,include_lowest=True)
per_school_summary
```

```python
# Calculate averages for the desired columns. 
size_math_scores = per_school_summary.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["% Passing Math"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["Overall Passing %"].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = {"Average Math Score": size_math_scores,
                    "Average Reading Score": size_reading_scores,
                    "% Passing Math": size_passing_math,
                    "% Passing Reading": size_passing_reading,
                    "Overall Passing %": size_overall_passing}

size_summary_df = pd.DataFrame(size_summary)

# Display results
size_summary_df
```
